# 1. Import Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 2. Loading the dataset into Pandas dataframe and replacing 'Not Given' data with NaN values

```python
df= pd.read_csv('dataset - netflix1_internship.csv', na_values='Not Given')
df
```

```python
df.isnull().sum()
```

# 3. Listing out the unique data values in each Category

```python
unique_type= df['type'].unique()
unique_country= df['country'].unique()
unique_year= df['release_year'].unique()
unique_rating= df['rating'].unique()

print("\nShow type:\n",unique_type)
print("\nCountries of release:\n",unique_country)
print("\nRelease years:\n",unique_year)
print("\nAge rating:\n",unique_rating)
```

# 4. Comparing the count of TV shows to Movies- Bar Plot

```python
counts= df['type'].value_counts()
counts

counts.plot(kind='bar', figsize=(8,6))
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Comparision of no. of TV Shows to Movies')
plt.show()
```

# 5. Timeline of the number of Releases per annum- Area Plot

```python
release_years= df['release_year'].value_counts()
release_years
release_years.sort_index()

release_years.plot(kind='area', figsize=(12,6))
plt.yticks(range(0,1500,100))
plt.xticks(range(1930, 2020, 5))
plt.xlabel('Year of Release')
plt.ylabel('Number of Shows')
plt.title('Timeline of Releases per Annum')
plt.show()
```

# 6. Releases Trend Analysis (Movies VS TV Shows)- Line Plot Comparision

```python

movies_release = df[df['type']== 'Movie']
tv_shows_release = df[df['type']== 'TV Show']

movies_release_years = movies_release['release_year'].value_counts().sort_index()
tv_shows_release_years = tv_shows_release['release_year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(movies_release_years.index, movies_release_years.values, color='b', label='Movies')
plt.plot(tv_shows_release_years.index, tv_shows_release_years.values, color='g', label='TV Shows')
plt.xlabel('Year of Release')
plt.ylabel('Number of Shows')
plt.title('Releases Trend Analysis- Movies v/s TV Shows')
plt.xticks(range(1930, 2020, 5))
plt.yticks(range(0, 1500, 100))
plt.legend()

plt.show()
```

# 7. Frequency Distribution of number of Releases per Country- Histogram

```python
countries= df['country'].dropna().astype(str)
countries

plt.figure(figsize=(30,10))
plt.hist(countries,bins=len(countries))
plt.xlabel('Country')
plt.ylabel('No. of Shows')
plt.title('Frequency Distribution of Number of Releases per Country', fontsize=30)
plt.xticks(rotation=45, ha='right')  
plt.grid(axis='y', linestyle='--')
plt.show()
```

# 8. Rating Distribution of Shows- Pie Plot

```python
ratings= df['rating'].value_counts()
print(ratings)

plt.figure(figsize=(5,5))
plt.pie(ratings, labels=ratings.index, autopct='%1.1f%%')
plt.title('Rating Distribution of Shows', fontsize= 16)
plt.show()
```

# 9. Analysis of Ratings among Multiple Genres

```python
rating_counts = {}
for genres, rating in zip(df['listed_in'].str.split(', '), df['rating']):
    for genre in genres:
        genre = genre.strip()  
        if genre not in rating_counts:
            rating_counts[genre] = {}
        rating_counts[genre][rating] = rating_counts[genre].get(rating, 0) + 1

rating_on_genre = pd.DataFrame(rating_counts).T.fillna(0).astype(int)
rating_on_genre
```
